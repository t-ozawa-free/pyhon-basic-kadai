```python
def trapezoid_cal(top, bottom, height):
    """
    台形の面積を計算する関数
    top: 上辺の長さ
    bottom: 下辺の長さ
    height: 高さ
    """
    return (top + bottom) * height / 2

top = 10  # 上辺：10cm
bottom = 20  # 下辺：20cm
height = 5  # 高さ：5cm
    
area = trapezoid_cal(top, bottom, height)  # 関数を使用して台形の面積を計算
print(f"台形の面積は{area}cm²です。")  # 計算した台形の面積を出力
```
