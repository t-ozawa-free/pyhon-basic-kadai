```python
from datetime import datetime

class Event:
    def __init__(self, name, date, location):
        self.name = name
        self.date = datetime.strptime(date, "%Y/%m/%d")
        self.location = location
    
    def cal_date(self):
        return (self.date - datetime.now()).days

event = Event("プログラミング初心者向けセミナー", "2026/05/30", "東京国際フォーラム")
print(event.name)
print(event.date)
print(event.location)
print(f"今日（{datetime.now()}）から、{event.cal_date()}日後に開催されます。")
```
