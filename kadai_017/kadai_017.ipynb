```python
human_dict = {"山田太郎": 40, "佐藤次郎": 30, "鈴木三郎": 20, "侍花子": 15}

class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}は大人です。")
        else:
            print(f"{self.name}は大人ではありません。")

# Humanの情報を保持した辞書から、インスタンスを生成し、リスト化して返す
def human_list_proc(human_dict):
    human_list = []
    # インスタンスの作成
    for name, age in human_dict.items():
        human = Human(name, age)
        human_list.append(human)
    return human_list

result_list = human_list_proc(human_dict)
# リストの分だけメソッドを呼び出す
for human in result_list:
    human.check_adult()

```
